```python
def rotl32(x, n):
    return ((x << n) & 0xffffffff) | (x >> (32 - n))


def quarter_round(a, b, c, d):
    a = (a + b) & 0xffffffff
    d ^= a
    d = rotl32(d, 16)

    c = (c + d) & 0xffffffff
    b ^= c
    b = rotl32(b, 12)

    a = (a + b) & 0xffffffff
    d ^= a
    d = rotl32(d, 8)

    c = (c + d) & 0xffffffff
    b ^= c
    b = rotl32(b, 7)
    return a, b, c, d


print('helper functions loaded')
```
