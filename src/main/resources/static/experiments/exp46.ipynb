### Step 1: 理解CFB模式核心工作原理

CFB（Cipher Feedback，密文反馈）是将分组密码转换为流密码使用的经典模式，核心逻辑是**用密文作为反馈生成密钥流**：
1. 初始化：生成随机初始化向量（IV），长度与分组长度一致（如16字节）；
2. 加密流程：
   a. 第一步：加密IV得到密钥流$E_K(IV)$，与第一个明文块$P_1$异或生成密文块$C_1=P_1⊕E_K(IV)$；
   b. 第二步：加密$C_1$得到密钥流$E_K(C_1)$，与第二个明文块$P_2$异或生成$C_2=P_2⊕E_K(C_1)$；
   c. 后续分组：依次用前一个密文块加密生成密钥流，与当前明文块异或得到密文块；
3. 解密流程：
   a. 第一步：加密IV得到$E_K(IV)$，与$C_1$异或还原$P_1=C_1⊕E_K(IV)$；
   b. 第二步：加密$C_1$得到$E_K(C_1)$，与$C_2$异或还原$P_2=C_2⊕E_K(C_1)$；
   c. 后续分组：与加密流程完全相同，仅将密文块与密钥流异或得到明文。
该模式的核心特征是：加解密使用完全相同的逻辑，仅通过异或操作完成，无需单独的解密函数。


> 提示：CFB的本质是“分组密码流密码化”：将分组密码生成的密钥流与明文逐位/逐块异或，兼具分组密码的安全性和流密码的灵活性

### Step 2: 分析CFB模式的实现特性

基于CFB的密文反馈原理，其实现层面有四个核心特性：
1. 必须使用IV：IV是生成初始密钥流的基础，需随机生成且长度与分组一致，重复使用IV+密钥会泄露明文；
2. 加密串行化：每个分组的密钥流依赖前一个密文块，加密必须按顺序执行，无法并行；
3. 解密串行化：解密同样依赖前一个密文块生成密钥流，因此也无法并行处理；
4. 有限错误传播：单个密文块的错误仅影响自身和下一个明文块的解密结果，后续分组可自动恢复（自同步特性）；
5. 无需填充：CFB直接处理任意长度数据，无需PKCS#7等填充方式（流密码特性）。
数学表达式上，CFB的加解密核心公式统一为：
加密：$C_i = P_i ⊕ E_K(C_{i-1}), C_0=IV$
解密：$P_i = C_i ⊕ E_K(C_{i-1})$
其中$C_0$固定为IV，加解密仅需加密函数$E_K$，无需解密函数$D_K$。


> 提示：CFB无需填充是重要优势：避免了padding oracle等针对填充的攻击，也简化了长度非分组整数倍数据的处理

### Step 3: 掌握CFB模式的安全优势

CFB模式结合了分组密码和流密码的优势，具备独特的安全特性：
1. 自同步能力：若传输过程中丢失/篡改部分密文块，错误仅影响有限个分组，后续分组可通过密文反馈自动恢复同步，适合不可靠的传输链路；
2. 任意长度支持：可处理任意长度的明文/密文，无需填充，适配流式数据（如字符流、实时通信数据）；
3. 语义安全性：IV的随机性保证相同明文生成不同密文，避免ECB模式的模式泄露问题；
4. 加解密逻辑统一：加解密使用完全相同的代码逻辑，仅需实现加密函数，简化开发和维护；
5. 抗篡改检测：密文块的篡改会导致对应明文块错误，且错误传播有限，便于定位问题。
CFB的主要安全风险与CBC一致：IV需随机生成，避免重复使用相同IV+密钥组合。


> 提示：CFB的自同步特性使其特别适合串口、终端等交互式通信场景，即使传输中出现少量错误也不会导致整体解密失败

### Step 4: 明确CFB模式的适用场景

CFB模式凭借流式加密和自同步的特性，主要适用于需要实时处理、容错性要求高的场景：
1. 字符流加密：终端输入、串口通信、键盘输入等逐字符加密场景；
2. 交互式应用：SSH、Telnet等远程登录协议的加密层，实时响应且容错性要求高；
3. 网络实时通信：VoIP、视频通话等实时流媒体的加密，允许少量错误但需快速恢复；
4. 不可靠传输链路：卫星通信、无线通信等易出现传输错误的场景，利用自同步特性减少重传；
5. 短消息加密：短信、即时消息等长度不固定的短数据加密，无需填充更高效。
核心适用条件：需要实时处理、流式传输、对错误传播有容忍度的场景。


> 提示：在需要高并行性的批量数据加密场景（如文件加密），CFB不如CBC/CTR高效，需根据场景选择

### Step 5: 实现CFB模式的加解密类

基于Python实现CFB模式的核心加解密逻辑，继承分组密码模式基类：
1. 初始化方法：指定分组长度（默认16字节），设置模式名称为"CFB"；
2. 加密方法encrypt：
   a. 处理IV：未提供则随机生成，提供则校验长度是否为分组长度；
   b. 初始化反馈值为IV，将IV拼接在密文开头（便于解密）；
   c. 按分组长度遍历明文（无需填充）：
      - 加密反馈值生成密钥流块；
      - 截取与当前明文块长度匹配的密钥流；
      - 明文块与密钥流异或生成密文块；
      - 拼接密文块，更新反馈值（完整块直接替换，不完整块拼接旧反馈值）；
3. 解密方法decrypt：
   a. 提取IV：未提供则从密文开头截取，提供则直接使用；
   b. 初始化反馈值为IV，按分组长度遍历密文：
      - 加密反馈值生成密钥流块；
      - 截取与当前密文块长度匹配的密钥流；
      - 密文块与密钥流异或生成明文块；
      - 拼接明文块，更新反馈值（规则同加密）；
   c. 直接返回明文（无需去填充）。

```python
class CFBMode(BlockCipherMode):
    """密文反馈模式 (Cipher Feedback Mode)"""
    
    def __init__(self, block_size: int = 16):
        super().__init__(block_size)
        self.name = "CFB"
    
    def encrypt(self, plaintext: bytes, key: bytes, iv: Optional[bytes] = None) -> bytes:
        cipher = SimpleAES(key)
        # 生成或使用提供的IV
        if iv is None:
            iv = self._generate_iv()
        elif len(iv) != self.block_size:
            raise ValueError(f"IV长度必须为{self.block_size}字节")
        # CFB模式不需要填充
        ciphertext = iv  # 将IV放在密文前面
        feedback = iv
        # 按分组大小处理明文
        for i in range(0, len(plaintext), self.block_size):
            # 加密反馈值
            keystream_block = cipher.encrypt_block(feedback)
            # 获取当前明文分组
            plaintext_block = plaintext[i:i + self.block_size]
            # 异或生成密文
            ciphertext_block = self._xor_blocks(plaintext_block, keystream_block[:len(plaintext_block)])
            ciphertext += ciphertext_block
            # 更新反馈值（使用密文作为反馈）
            if len(ciphertext_block) == self.block_size:
                feedback = ciphertext_block
            else:
                # 如果最后一个分组不完整，需要特殊处理
                feedback = ciphertext_block + feedback[len(ciphertext_block):]
        return ciphertext
    
    def decrypt(self, ciphertext: bytes, key: bytes, iv: Optional[bytes] = None) -> bytes:
        cipher = SimpleAES(key)
        if iv is None:
            # 从密文中提取IV
            if len(ciphertext) < self.block_size:
                raise ValueError("密文太短，无法包含IV")
            iv = ciphertext[:self.block_size]
            actual_ciphertext = ciphertext[self.block_size:]
        else:
            actual_ciphertext = ciphertext
        plaintext = b''
        feedback = iv
        # 按分组大小处理密文
        for i in range(0, len(actual_ciphertext), self.block_size):
            # 加密反馈值
            keystream_block = cipher.encrypt_block(feedback)
            # 获取当前密文分组
            ciphertext_block = actual_ciphertext[i:i + self.block_size]
            # 异或生成明文
            plaintext_block = self._xor_blocks(ciphertext_block, keystream_block[:len(ciphertext_block)])
            plaintext += plaintext_block
            # 更新反馈值（使用密文作为反馈）
            if len(ciphertext_block) == self.block_size:
                feedback = ciphertext_block
            else:
                # 如果最后一个分组不完整，需要特殊处理
                feedback = ciphertext_block + feedback[len(ciphertext_block):]
        return plaintext
```

## 小实验：敏感性（雪崩效应）

```python
def hamming_distance_bytes(a: bytes, b: bytes) -> int:
    if len(a) != len(b):
        raise ValueError("长度必须相同才能计算差异度")
    return sum(x != y for x, y in zip(a, b))

def flip_one_bit(data: bytes, bit_index: int = 0) -> bytes:
    if not data:
        return data
    byte_i = bit_index // 8
    bit_i = bit_index % 8
    byte_i = byte_i % len(data)
    mask = 1 << bit_i
    arr = bytearray(data)
    arr[byte_i] ^= mask
    return bytes(arr)

def core_transform(data: bytes) -> bytes:
    # TODO: 替换成你的核心变换
    # 示例：用 SHA-256 作为“对照组”
    return hashlib.sha256(data).digest()

base = b"hello world"
out1 = core_transform(base)
out2 = core_transform(flip_one_bit(base, 0))

print("输出长度(字节):", len(out1))  # 预期输出：32（若使用SHA-256）
print("差异度(字节):", hamming_distance_bytes(out1, out2))  # 预期输出：通常 > 10
```

## 附录A：最常用的数学与位运算背景

### 1. 模运算（Modular Arithmetic）

当我们写 $$a \equiv b \pmod n$$ 时，意思是 $a$ 与 $b$ 除以 $n$ 的余数相同，也就是 $n$ 整除 $a-b$。

常见等价写法：

- $a \bmod n$：把 $a$ 映射到 $0..n-1$ 的代表元
- 若 $a<0$，Python 仍保证 $a \bmod n \in [0,n-1]$

**一个很重要的直觉：** 模运算会“折叠”整数轴，把无限多的整数映射到有限集合，所以很多密码体制会用到它来构造“封闭”的运算空间。

### 2. 最大公因数与互质

若 $$\gcd(a,n)=1$$，我们称 $a$ 与 $n$ 互质（coprime）。互质非常重要，因为它意味着 $a$ 在模 $n$ 意义下存在乘法逆元 $a^{-1}$，满足：

$$a\cdot a^{-1} \equiv 1 \pmod n$$

### 3. 扩展欧几里得算法（Extended Euclid）

它不仅能算 $\gcd(a,b)$，还能找到整数 $x,y$ 使得：

$$ax+by=\gcd(a,b)$$

当 $\gcd(a,n)=1$ 时，$x$ 就是 $a$ 关于模 $n$ 的逆元（可能为负，需要再取模）。

### 4. 异或（XOR）

在字节层面，异或常写成 $c=a\oplus b$，它有几个极其重要的性质：

- $a\oplus a=0$
- $a\oplus 0=a$
- $(a\oplus b)\oplus b=a$（可逆性）

因此很多流密码/分组模式会用异或把“密钥流”与明文混合。

> 调试提示：如果你发现解密结果不等于明文，先检查“同一份密钥流是否被一致地使用”，以及字节对齐是否正确。


```python
# 附录A配套代码：gcd、逆元、异或操作的最小实现与自测

def egcd(a: int, b: int) -> Tuple[int, int, int]:
    """返回 (g, x, y) 使得 a*x + b*y = g = gcd(a,b)"""
    if b == 0:
        return (a, 1, 0)
    g, x1, y1 = egcd(b, a % b)
    x = y1
    y = x1 - (a // b) * y1
    return (g, x, y)

def modinv(a: int, n: int) -> int:
    g, x, _ = egcd(a, n)
    if g != 1:
        raise ValueError("不存在逆元：a 与 n 不互质")
    return x % n

print(math.gcd(3, 26))     # 预期输出：1
print(modinv(3, 26))       # 预期输出：9，因为 3*9=27≡1(mod26)

def xor_bytes(a: bytes, b: bytes) -> bytes:
    if len(a) != len(b):
        raise ValueError("xor 需要等长字节串")
    return bytes(x ^ y for x, y in zip(a, b))

p = b"ABC"
k = b"\x01\x01\x01"
c = xor_bytes(p, k)
p2 = xor_bytes(c, k)
print(c)   # 预期输出：b'@BA'（因为 0x41^1=0x40 等）
print(p2)  # 预期输出：b'ABC'
```
